```python
# #공식############################
# CHNNL_RE_ALL
# CHNNL_RE_1M
# CHNNL_RE_3M
# CHNNL_RE_P_VIDO
# CHHHL_RE_P_V


# (likeCount-dislikeCount)*commentCount
# (likeCount_1M-dislikeCount_1M)*commentCount_1M
# (likeCount_3M-dislikeCount_3M)*commentCount_3M
# commentCount/videoCount
# commentCount/((collectedTime-publishedAt)/7)
```

```python
o_CHNNL_RE_ALL = (orig_df["likeCount"]-orig_df["dislikeCount"])*orig_df["commentCount"]
# CHNNL_RE_1M 1달치이므로 생략 (CHNNL_RE_ALL과 같은값)
# CHNNL_RE_3M 1달치이므로 생략 (CHNNL_RE_ALL과 같은값)
o_CHNNL_RE_P_VIDO = (orig_df["commentCount"]/orig_df["videoCount"])


# CHHHL_RE_P_V:  commentCount/((collectedTime-publishedAt)/7)
# groupby형태로 채널의 video들을 sum하여 index 데이터를 가공하여 생성,  채널의 collectedTime, publishedAt 값을 알 수 없음
# 그러나 export에는 collectedTime, publishedAt 존재
o_CHNNL_RE_P_VIDO
```

```python
o_CHNNL_RE_ALL
min0 = min(o_CHNNL_RE_ALL.unique())
max0 = max(o_CHNNL_RE_ALL.unique())
o_CHNNL_RE_ALL_100 = ((o_CHNNL_RE_ALL- min0)/(max0-min0))*100
```

```python
o_CHNNL_RE_P_VIDO
min1 = min(o_CHNNL_RE_P_VIDO.unique())
max1 = max(o_CHNNL_RE_P_VIDO.unique())
o_CHNNL_RE_P_VIDO_100 = ((o_CHNNL_RE_P_VIDO- min1)/(max1-min1))*100
```

```python
o_CHNNL_RE_ALL_ZSC = ss.zscore(o_CHNNL_RE_ALL)
o_CHNNL_RE_P_VIDO_ZSC = ss.zscore(o_CHNNL_RE_P_VIDO)
```
